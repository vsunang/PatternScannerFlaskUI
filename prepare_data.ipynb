```python
import pandas as pd
import tushare as ts
from datetime import datetime, timedelta
```

```python
today = datetime.today().strftime('%Y%m%d')
half_year_before = ( datetime.today() - timedelta(days=180) ).strftime('%Y%m%d')
tickers = pd.read_csv('data800_list.csv')
```

```python
for code in tickers['code']:
    print(code)
    df = ts.pro_bar(ts_code=code, adj='qfq', start_date=half_year_before, end_date=today)
    df.to_csv(f'./data800/{code}.csv')
```
